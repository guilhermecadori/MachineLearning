```python
# Importing supporting libs
import numpy as np
import matplotlib.pyplot as plt

# Input is an array
input_array = np.array([1, 2, 3])
exp_array = np.exp([input_array])

# Implementation of the sigmoid function (logistic funtion)
def sigmoid(Z):
    # Imports supporting library if this fn. is reused without previously calling 'numpy'
    import numpy as np
    
    g = 1 / (1 + np.exp(-Z))
    
    return g

# Testing the implemented function
# Generating data - evenly spaced values between -10 and 10
z_test = np.arange(-10, 11)

# Calling the implemented function
y = sigmoid(z_test)

# Ploting the function
# Creating graph objecy
fig, ax = plt.subplots(1, 1, figsize=(5, 3))
ax.plot(z_test, y, c='b')

# Classification threshold is y=0.5 --> x=0
ax.vlines(x=0, ymin=min(y), ymax=max(y), colors='r', label='x=0')

# Creating graph title and axis labels
ax.set_title('Implemented Sigmoid Function')
ax.set_ylabel('Sigmoid (z)')
ax.set_xlabel('z')

# Plotting the graph
plt.show()
```

```python
# Creating data
x = np.array([[0.5, 1.5], [1, 1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1, 1)

# Plotting data
fig, ax = plt.subplots(1, 1, figsize=(4, 4))

# Graph configs
ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')

# Configuring values to be plotted
for i in range(len(y)):
    if y[i] == 0:
        ax.plot(x[i, 0], x[i, 1], 'bo')
    else:
        ax.plot(x[i, 0], x[i, 1], 'rx')
        
        
# Choose values between 0 and 6
# Logistic Regression model is: f(x) = g(x0 + x1 - 3)
# If x0 + x1 - 3 >= 0, then y = 1, else y = 0
# Plotting x0 + x1 - 3 >= 0, which is equivalent to x1 = 3 - x0

# Setting values for x0
x0 = np.arange(0,6)

# Calculating values for x1
x1 = 3 - x0

# Plot the decision boundary
ax.plot(x0, x1, c="b")
ax.axis([0, 4, 0, 3.5])

# Fill the region below the line
ax.fill_between(x0,x1, alpha=0.2)

# Adding a legend to the graph
# Plotting blue 'O' for y=0
ax.plot(x[y.flatten() == 0, 0], x[y.flatten() == 0, 1], 'bo', label='y=0')  # Blue 'O' for y=0

# Plotting red 'X' for y=1
ax.plot(x[y.flatten() == 1, 0], x[y.flatten() == 1, 1], 'rx', label='y=1')  # Red 'X' for y=1

# Adding custom legend
ax.legend()


# Plotting graph
plt.show()
```
